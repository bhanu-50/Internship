```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sqlite3
warnings.filterwarnings('ignore')
```

```
Customer = pd.read_csv('/content/Internship/Datasets/Customers.csv',encoding='latin-1',sep=';')
Customer.head()
```

```
Locations = pd.read_csv('/content/Internship/Datasets/Location.csv',encoding='latin-1',sep=';')
Locations.head()
```

```
Orders = pd.read_csv('/content/Internship/Datasets/Orders.csv',encoding='latin-1',sep=';')
Orders.head()
```

```
Products = pd.read_csv('/content/Internship/Datasets/Products.csv',encoding='latin-1',sep=';')
Products.head()
```

```
datasets = [Customer,Locations,Orders,Products]
Names = ['Customer','Locations','Orders','Products']
```

```
for i in range(len(datasets)):
  print("="*50)
  print("\n"+"Dataset: "+ Names[i]+"\n")
  print(datasets[i].info())
```

```
# Join Customer and Orders on 'Customer ID'
df_merged = pd.merge(Customer, Orders, on='Customer ID', how='inner')

# Join the result with Products on 'Product ID'
df_merged = pd.merge(df_merged, Products, on='Product ID', how='inner')

# Join the result with Locations on 'Postal Code'
# Note: 'Postal Code' is float in Locations and int in Orders, we need to convert one to match
df_merged['Postal Code'] = df_merged['Postal Code'].astype(float)
Locations['Postal Code'] = Locations['Postal Code'].astype(float)
df_merged = pd.merge(df_merged, Locations, on='Postal Code', how='inner')


# Display the first few rows of the final merged DataFrame
display(df_merged.head())
```
